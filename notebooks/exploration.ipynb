# Cadmium Lake Exploration

This notebook provides a minimal reproducible exploration surface for layer distributions, cross-layer variability, coverage, and traceability drilldown.

```python
import duckdb
import pandas as pd
from pathlib import Path

db_path = Path('../data/curated/cadmium_lake.duckdb')
conn = duckdb.connect(str(db_path))
```

```python
layer_distribution = conn.execute('select * from layer_distribution_view').df()
layer_distribution.head()
```

```python
coverage = conn.execute('select * from source_coverage_view').df()
coverage
```

```python
traceability = conn.execute('''
select n.measurement_id, n.canonical_value, n.canonical_unit, r.raw_value_text, r.raw_unit, r.page_or_sheet, r.table_or_figure, s.sample_name, s.matrix_group, sf.local_path
from measurements_normalized n
join measurements_raw r using (measurement_id)
join samples s using (sample_id)
left join source_files sf on sf.source_id = s.source_id
limit 20
''').df()
traceability
```
